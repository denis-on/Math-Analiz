### 1.Найти интервалы возрастания и убывания функций

* Чтобы найти интервалы возрастания и убывания функции f(x) = x + e^(-x), можно использовать производную функции.

Сначала найдем производную функции:

f'(x) = 1 - e^(-x)

Затем приравняем ее к нулю, чтобы найти критические точки:

f'(x) = 0

1 - e^(-x) = 0

e^(-x) = 1

x = 0

Теперь выясним знак производной для каждого интервала:

(-бесконечность, 0): f'(x) < 0. Функция убывает.
(0, +бесконечность): f'(x) > 0. Функция возрастает.
Таким образом, интервалы возрастания и убывания функции f(x) = x + e^(-x) следующие:

Функция убывает на интервале (-бесконечность, 0).
Функция возрастает на интервале (0, +бесконечность).

* Чтобы найти интервалы возрастания и убывания функции f(x) = x * ln(x), можно использовать производную функции.

Сначала найдем производную функции:

f'(x) = 1 * ln(x) + x * (1/x) = ln(x) + 1

Затем приравняем ее к нулю, чтобы найти критические точки:

f'(x) = 0

ln(x) + 1 = 0

ln(x) = -1

x = e^-1 = 1/e

Теперь выясним знак производной для каждого интервала:

(-бесконечность, 0): f'(x) < 0. Функция убывает.
(0, 1/e): f'(x) > 0. Функция возрастает.
(1/e, +бесконечность): f'(x) > 0. Функция возрастает.
Таким образом, интервалы возрастания и убывания функции f(x) = x * ln(x) следующие:

Функция убывает на интервале (-бесконечность, 0).
Функция возрастает на интервалах (0, 1/e) и (1/e, +бесконечность).

* Для функции f(x) = 1/(1 - x^2) найдем производную функции:

f(x) = (1 - x^2)^(-1)

f'(x) = -(2x)(1 - x^2)^(-2)

Затем приравняем ее к нулю, чтобы найти критические точки:

f'(x) = 0

-(2x)(1 - x^2)^(-2) = 0

x = 0 x = 1

Далее, для каждого интервала требуется определить знак производной:

(-бесконечность, -1): f'(x) < 0. Функция убывает.
(-1, 0): f'(x) > 0. Функция возрастает.
(0, 1): f'(x) > 0. Функция возрастает.
(1, +бесконечность): f'(x) < 0. Функция убывает.
Таким образом, интервалы возрастания и убывания функции f(x) = 1/(1 - x^2) следующие:

Функция убывает на интервалах (-бесконечность, -1) и (1, +бесконечность).
Функция возрастает на интервалах (-1, 0) и (0, 1).

### 3. Найти интервалы выпуклости и точки перегиба функций

Чтобы найти интервалы выпуклости и точки перегиба функции, мы можем использовать библиотеку SciPy для вычисления производных и вторых производных.

```python
# функция для вычисления второй производной
def second_derivative(f, x):
     return derivative(f, x, n=2)
```

```python
# задаем функцию y=cos(x)
f = lambda x: np.cos(x)

# находим точки перегиба
x_values = np.linspace(-1.5*np.pi, 1.5*np.pi, 100)
second_derivatives = [second_derivative(f, x) for x in x_values]
inflection_points = [(x_values[i], x_values[i+1]) for i in range(len(second_derivatives)-1)
                     if np.sign(second_derivatives[i]) != np.sign(second_derivatives[i+1])]

# находим интервалы выпуклости и вогнутости
convex_intervals = []
concave_intervals = []
for i in range(1, len(x_values)-1):
    second_derivative_value = second_derivative(f, x_values[i])
    if second_derivative_value > 0:
        convex_intervals.append((x_values[i-1], x_values[i]))
    elif second_derivative_value < 0:
        concave_intervals.append((x_values[i-1], x_values[i]))

#print("Интервалы выпуклости:", convex_intervals)
#print("Интервалы вогнутости:", concave_intervals)
print("Точки перегиба:", inflection_points)
```

```python
plt.plot(x_values, f(x_values))
for interval in convex_intervals:
    plt.axvspan(interval[0], interval[1], alpha=0.2, color='green')
for interval in concave_intervals:
    plt.axvspan(interval[0], interval[1], alpha=0.2, color='red')
for point in inflection_points:
    plt.axvline(x=point[0], linestyle='--', color='blue')
plt.show()
```

### 5. Провести полное исследование и построить графики функций

```python
# Определение функции
def f(x):
    return np.log(1 - x ** 2)

# Создание массива значений аргумента x
x_values = np.linspace(-1, 1, 1000)

# Вычисление значений функции для каждого значения x
y_values = f(x_values)

# Построение графика функции
plt.plot(x_values, y_values, label='y=ln(1-x^2)')

# Добавление заголовка и меток осей
plt.title('График функции y=ln(1-x^2)')
plt.xlabel('x')
plt.ylabel('y')

# Нахождение точек пересечения с осями координат
plt.plot([0, 0], [np.min(y_values), np.max(y_values)], linestyle='--', color='gray', label='ось X')

# Добавление сетки на график
plt.grid(True)

# Нахождение интервалов возрастания и убывания функции
plt.plot([-1, -1], [np.min(y_values), np.max(y_values)], linestyle=':', color='gray')
plt.plot([0, 0], [np.min(y_values), np.max(y_values)], linestyle=':', color='gray')
plt.plot([1, 1], [np.min(y_values), np.max(y_values)], linestyle=':', color='gray')

# Нахождение экстремумов функции
plt.plot([0], [f(0)], marker='o', color='red', markersize=5, label='точка максимума')

# Добавление горизонтальной асимптоты
plt.plot([np.min(x_values), np.max(x_values)], [-np.inf, -np.inf], linestyle='--', color='gray', label='горизонтальная асимптота')

# Добавление легенды и вывод графика
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Определение функции
def f(x):
    return x ** 2 / (1 - (x ** 2))

# Создание массива значений аргумента x
x_values = np.linspace(-1, 1, 100)

# Вычисление значений функции для каждого значения x
y_values = f(x_values)

# настройка масштаба оси y
plt.ylim(-1, 25) 

# Построение графика функции
plt.plot(x_values, y_values, label='y=x^2/(1-x^2)')

# Добавление заголовка и меток осей
plt.title('График функции y=x^2/(1-x^2)')
plt.xlabel('x')
plt.ylabel('y')

# Добавление сетки на график
plt.grid(True)

# Нахождение области определения функции
x_domain = np.delete(x_values, np.where(np.abs(1 - x_values ** 2) < 1e-10))
plt.plot(x_domain, np.zeros_like(x_domain), linestyle='--', color='green', label='область определения')

# Нахождение точек пересечения с осями координат
plt.plot([0, 0], [np.min(y_values), np.max(y_values)], marker='x', linestyle='--', color='gray', label='ось X')
plt.plot([np.min(x_values), np.max(x_values)], [0, 0], linestyle='--', color='gray', label='ось Y')

# Нахождение интервалов возрастания и убывания функции
plt.plot([-1, -1], [np.min(y_values), np.max(y_values)], linestyle=':', color='gray')
plt.plot([0, 0], [np.min(y_values), np.max(y_values)], linestyle=':', color='gray')
plt.plot([1, 1], [np.min(y_values), np.max(y_values)], linestyle=':', color='gray')

# Нахождение экстремумов функции
derivative = (2*x_values) / ((1-x_values**2)**2)
extrema_idx = np.where(np.diff(np.sign(derivative)))[0] + 1
plt.plot(x_values[extrema_idx], y_values[extrema_idx], marker='o', color='red', markersize=5, label='точки экстремума')

# Добавление легенды и вывод графика
plt.legend()
plt.show()
```

Теперь мы можем провести исследование функции. Для начала определим, что функция является положительной на всей числовой прямой.

```python
x = np.linspace(-10, 10, 500)
y = func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x^2*e^(-x)')
plt.grid()
plt.show()
```

Функция имеет максимальное значение при x=2, после чего она начинает убывать. Также мы можем заметить, что функция изменяет свой знак в точке x=0, что означает, что это точка пересечения графика с осью x. Далее, мы можем вычислить первую и вторую производные функции:

```python
def first_derivative(x):
    return 2*x*np.exp(-x) - x**2*np.exp(-x)

def second_derivative(x):
    return 2*np.exp(-x) - 4*x*np.exp(-x) + x**2*np.exp(-x)

x = np.linspace(0, 10, 500)

plt.subplot(2, 1, 1)
plt.plot(x, func(x), label='y=x^2*e^(-x)')
plt.plot(x, first_derivative(x), label="y'=2xe^(-x)-x^2e^(-x)")
plt.xlabel('x')
plt.grid()
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, second_derivative(x), label="y''=2e^(-x)-4xe^(-x)+x^2e^(-x)")
plt.xlabel('x')
plt.grid()
plt.legend()

plt.show()
```

Из графиков производных мы можем сделать следующие выводы:

Первая производная равна нулю в точке $x=2$, что означает, что это точка максимума функции.
Первая производная положительна на всей числовой прямой, кроме точки пересечения с осью x, поэтому функция возрастает на всей числовой прямой, кроме положительной полуоси.
Вторая производная положительна на всей числовой прямой, кроме точки пересечения с осью x, что означает, что график функции выпуклый вверх на всей числовой прямой, кроме положительной полуоси.
Из общего описания функции и её исследования мы можем сделать следующие выводы:

Функция $y=x^2 e^{-x}$ является положительной на всей числовой прямой.
Функция имеет максимальное значение при $x=2$.
Функция уменьшается на всей числовой прямой после точки максимума $x=2$.
Функция пересекает ось x в точке $x=0$.
Функция является монотонно убывающей на отрицательной полуоси и монотонно возрастающей на положительной полуоси.
График функции выпуклый вверх на всей числовой прямой.

```python
def func(x):
    return x - np.log(x)

x = np.linspace(0.1, 10, 500) # избегаем возможные деления на 0 и отрицательные значения аргумента
y = func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=x-ln(x)')
plt.grid()
plt.show()
```

Из графика видно, что функция убывает со скоростью, которая уменьшается при увеличении аргумента $x$. На промежутке $(0, 1]$ функция отрицательна, а на $[1, +\infty)$ - положительна. Минимальное значение функция достигает в точке $x=1$. Для получения более подробной информации проведем исследование функции:

Для этой функции определен область определения $D=[0,+\infty)$, так как $\ln x$ определен только для $x>0$.
Функция является непрерывной на области определения, поскольку представляет собой разность непрерывных функций $x$ и $\ln x$.
Функция имеет наклонный асимптоту в точке $x=0$, поскольку $\lim_{x\to0+}x-\ln x=+\infty$.
В точке $x=1$ функция принимает минимальное значение $0$.
Функция не имеет экстремумов на промежутках $[0,1)$ и $(1,+\infty)$, так как первая производная $y'=1-\frac{1}{x}$ на них разных знаков (отрицательная на первом промежутке и положительная на втором).
Вторая производная $y''=\frac{1}{x^2}$ на всей области определения положительна, следовательно, функция является выпуклой вверх на всей области определения.
