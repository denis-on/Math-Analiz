Построить графики функций

```python
# Создаем массив точек x с шагом 0.1
x = np.arange(-10, 6, 0.1)

# Вычисляем функцию y(x) для каждой точки x
y = x*x + 4*x + 3
```

```python
# Строим график функции
plt.plot(x, y)

# Добавляем заголовок и подписи осей
plt.title("График функции y(x) = x^2 + 4x + 3")
plt.xlabel("x")
plt.ylabel("y")

# Показываем график
plt.show()
```

```python
# Создаем массив точек x с шагом 0.01
x = np.arange(-2.5, 2.5, 0.001)

# Вычисляем функцию y(x) для каждой точки x
y = (-2)*np.sin(3*x);
```

```python
# Строим график функции
plt.plot(x, y)

# Добавляем заголовок и подписи осей
plt.title("График функции y(x) = -2sin(3x)")
plt.xlabel("x")
plt.ylabel("y")

# Показываем график
plt.show()
```

```python
# Создаем массив точек x с шагом 0.01
x = np.arange(-2.5, 2.5, 0.001)

# Вычисляем функцию y(x) для каждой точки x
def fl(t):
    return t-np.floor(t)
y = abs(fl(x)-(1/2))
```

```python
# Строим график функции
plt.plot(x, y)

# Добавляем заголовок и подписи осей
plt.title("График функции y(x) = |{x} − 1/2|")
plt.xlabel("x")
plt.ylabel("y")

# Показываем график
plt.show()
```

Найти пределы:

lim(x^2−6x+5)/(x^2−25)\
x→5

```python
x = Symbol('x')
fun = (x**2-6*x+5)/(x**2-25)
limit(fun, x, 5)
```

lim(x^3+x+2)/(x^3+1)\
x→-1

```python
x = Symbol('x')
fun = (x**3+x+2)/(x**3+1)
limit(fun, x, -1)
```

lim(sqrt(2x+3)−3)/(sqrt(x−2)−1)\
x→3

```python
x = Symbol('x')
fun = (((2*x+3)**(1/2))-3)/(((x-2)**(1/2))-1)
limit(fun, x, 3)
```

lim(1−cos(x))/x^2\
x→0

```python
x = Symbol('x')
fun = (1-cos(x))/(x**2)
limit(fun, x, 0)
```

lim(x ctg(x))\
x→0

```python
x = Symbol('x')
fun = x/tan(x)
limit(fun, x, 0)
```

lim((1+3x)^(1/2x))\
x→0

```python
x = Symbol('x')
fun = (1+3*x)**(1/(2*x))
limit(fun, x, 0)
```

lim((3+5x)/(3+2x))^(1/x)
x→0

```python
x = Symbol('x')
fun = ((3+5*x)/(3+2*x))**(1/x)
limit(fun, x, 0)
```
